```python
import os
from supabase import create_client, Client

# url: str = os.environ.get("SUPABASE_URL")
url = "http://host.supabase:8020"
key: str = os.environ.get("SUPABASE_KEY")
supabase: Client = create_client(url, key)
response = (
    supabase
    .table("Item")
    .select("*")
    .execute()
)
print(response.data)
```
