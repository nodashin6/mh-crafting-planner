```python
class FileRouter:
    
    def __init__(self, dataset_id: str):
        self.dataset_id = dataset_id
    
    @property
    def data_dir(self):
        return DATA_DIR
    
    @property
    def dataset_dir(self):
        return DATASET_DIR
    
    def datetaset_yaml_path(self):
        return self.dataset_dir / f"{self.dataset_id}.yml"
```

```python
class DatasetYamlHandler:
    
    def __init__(self, dataset_id: str):
        self.dataset_id = dataset_id
        self.router = FileRouter(dataset_id)
        
    def load(self):
        data = self._load()
        
    def _load(self):
        with open(self.router.datetaset_yaml_path(), "r") as f:
            return yaml.safe_load(f)
```
